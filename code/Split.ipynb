```python
import pandas as pd
import numpy as np
```

```python
#reading csv file
data = pd.read_csv('final_data.csv')
from sklearn.utils import shuffle
#shuffling data
data = shuffle(data)
data.head()
```

```python
#function to split train and test data
def splitData(df, chunkSize = int(0.7*len(data))): 
    listOfDf = list()
    numberChunks = len(df) // chunkSize + 1
    for i in range(numberChunks):
        listOfDf.append(df[i*chunkSize:(i+1)*chunkSize])
    return listOfDf
```

```python
data = splitData(data)
#training data
data_train = data[0]
#testing data
data_test = data[1]

data_train.to_csv('data_final_train.csv', sep=',')
data_test.to_csv('data_final_test.csv', sep=',')
```
